# Kalman Filter



Time Update Equations: 

$$ \hat{x}^{-}_{k} = \hat{x}_{k-1}$$
$$ P^{-}_{k} = P_{k-1} + Q$$

Measurement Update Equations:

$$ K_{k} = P^{-}_{k} (P^{-}_{k} + R)^{-1} $$
$$ \hat{x}_{k} = \hat{x}^{-}_{k} + K_{k} (z_{k}-\hat{x}_{k}^{-})$$
$$ P_{k} = (1 - K_{k}) P^{-}_{k}$$

where $x$ is our estimate, $P^{-}$ our a priori error estimate, $P$ our a posteriori error estimate, $a$ the state transition (one if we observe directly, $z$ the measurements, $Q$ the transition error, $K$the kalman gain, $r$ the measurement error

### 1 Tracking a scalar Value


```python
import numpy as np
import matplotlib.pyplot as plt

# intial parameters
n_iter = 200
x = -0.4 


def KG_1D(n_iter, target, R, noise_measurement, Q, initial_guess, intial_error_estimate):
    x = target
    z = np.random.normal(x,noise_measurement,size=n_iter)
# allocate space for arrays
    xhat=np.zeros(n_iter)      # a posteri estimate of x
    P=np.zeros(n_iter)         # a posteri error estimate
    xhatminus=np.zeros(n_iter) # a priori estimate of x
    Pminus=np.zeros(n_iter)    # a priori error estimate
    K=np.zeros(n_iter)         # gain or blending factor

    # intial guesses
    xhat[0] = initial_guess#0.0
    P[0] = intial_error_estimate#1.0

    for k in range(1,n_iter):
        # time update
        xhatminus[k] = xhat[k-1]
        Pminus[k] = P[k-1]+Q

        # measurement update
        K[k] = Pminus[k]/( Pminus[k]+R )
        xhat[k] = xhatminus[k]+K[k]*(z[k]-xhatminus[k])
        P[k] = (1-K[k])*Pminus[k]
    return z, xhat, x , Pminus

z, xhat, x, Pminus  = KG_1D(200, -0.4, 0.1**2, 0.1, 10e-5, 0, 1)

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(z,'k+',label='noisy measurements')
ax.plot(xhat,'b-',label='a posteri estimate')
ax.axhline(x,color='g',label='truth value')
ax.set_title('Estimate vs. iteration step', fontweight='bold')
ax.set_xlabel('Iteration')
ax.set_ylabel('1D Quantity')
plt.legend()

fig, ax = plt.subplots(figsize=(15,8))
valid_iter = range(1,n_iter) # Pminus not valid at step 0
ax.plot(valid_iter,Pminus[valid_iter],label='a priori error estimate')
ax.set_title('Estimated $\it{\mathbf{a \ priori}}$ error vs. iteration step', fontweight='bold')
ax.set_xlabel('Iteration')
ax.set_ylabel('1D Quantity')
plt.setp(plt.gca(),'ylim',[0,.006])
plt.show()
```

### Now let's try to track a signal 

```python
x_axis = np.linspace(0, 5*np.pi, n_iter)
#KG_1D(n_iter, target, R, Q, initial_guess, intial_error_estimate)
z, xhat, x, Pminus  = KG_1D(200, 20*(np.sin(x_axis)), 0.1**2, 5, 10e-5, 0, 1)
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(x_axis, z,'k+',label='noisy measurements')
ax.plot(x_axis, xhat,'b-',label='a posteri estimate')
ax.plot(x_axis, x,color='g',label='truth value')
ax.set_title('Estimate vs. iteration step, signal tracker', fontweight='bold')
ax.set_xlabel('Iteration')
ax.set_ylabel('1D Quantity')
plt.legend()


fig, ax = plt.subplots(figsize=(15,8))
valid_iter = range(1,n_iter) # Pminus not valid at step 0
ax.plot(valid_iter,Pminus[valid_iter],label='a priori error estimate')
ax.set_title('Estimated $\it{\mathbf{a \ priori}}$ error vs. iteration step', fontweight='bold')
ax.set_xlabel('Iteration')
ax.set_ylabel('1D Quantity')
plt.setp(plt.gca(),'ylim',[0,.006])
plt.show()
```
